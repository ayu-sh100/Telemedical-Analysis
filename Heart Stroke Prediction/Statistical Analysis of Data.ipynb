```python
sns.countplot(x="HeartDisease",data=dataset)
```

```python
count = 0
for x in dataset.HeartDisease:
    if x == 1:
        count += 1
print(f"People having heart disease = {count}")
print(f"People not having heart disease = {len(dataset)-count}")
```

```python
sns.countplot(x="Smoking",hue="HeartDisease",data=dataset)
plt.legend(labels=['No','Yes'],title="Heart Disease?")
```
